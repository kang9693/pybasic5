###데이터 시각화?
> 데이터 분석 결과를 쉽게 이해할 수 있도록 시각적으로 표현하고 전달되는 과정을 말한다. 데이터 시각화의 목적은 도표(graph)라는 수단을 통해 정보를 명확하고 효과적으로 전달하는 것이다. - Wikipedia

![sheet](img/Picture1.png)

![bar](img/Picture2.png)

####시계열
* 시간에 따라 변하는 수치 데이터
* 트랜드를 확인할 때 유용
* 선 차트 등
    * 연간 실업률 추이
    * 연간 강수량 추이
    
![line](img/Picture3.png)

####등위
* 범주가 오름차순∙내림차순으로 순위가 매겨지는 경우
* 동일한 시점의 변수를 비교
* 막대 차트 등
    * 회사별 판매 실적

![bar](img/Picture4.png)

####비율
* 범주가 전체 대 개별 항목에 대한 비율인 경우
* 파이차트, 막대 차트 등
    * 시장 점유율 등
    
![Pie](img/Picture5.png)

#### 편차
* 기준과의 차이를 표시. 시간 흐름에 따른 변화가 주 대상
* 기준선이 중앙에 위치하여 값이 기준건 밑으로 표시되는 경우가 있다.
    * 예산안 대 지출 비교
    
![bar](img/Picture6.png)

####주기 분포
* 변수가 관찰된 개수를 일정 간격으로 표시
* 간격은 시간, 값의 범위 등이 될 수 있다.
    * 1시~2시, 2시~3시 등
    * 170~180cm, 180~190cm 등
* 히스토그램 등으로 표시

![histogram](img/Picture7.png)

####상관도
* 두 변수의 추세가 동일∙반대 방향인지 표시
* 산점도 등
    * 물가상승률과 실업률 관계

![scatter](img/Picture8.png)

####명목 속성 비교
* 범주가 순서를 매길 수 없는 경우 값을 비교
    * 상품 별 판매량
    * 가계부

![bar](img/Picture9.png)

####지리 정보
* 지역, 지리 별 값을 비교할 때 지도를 활용
* 지역 크기 변형, 색, 다른 시각화 레이어 추가 등
    * 변량비례도(Cartogram)
    * 계급구분도(Choropleth)
    
![cartogram](img/Picture10.png)

###Bokeh
#### Python 으로 웹브러우저 상의 인터랙티브 시각화를 생성하는 라이브러리
* low level, intermediate-level, high-level 인터페이스 제공
* [Intermediate-level 제공 시각화 종류](http://bokeh.pydata.org/en/latest/docs/reference/plotting.html)
* [Gallery](http://bokeh.pydata.org/en/latest/docs/gallery.html)

### D3.js
#### Data-Driven Documents
* DOM 요소에 데이터를 연결, 제어를 돕는 Javascript 라이브러리
* svg 만을 위한 라이브러리는 아님
    * p, div, span 태그 등 모든 태그에 적용 가능
    * svg 제어에 특히 유용
* 많은 플러그인
    * 동적 시각화 제작에 유용
    * [D3.js Gallery](https://github.com/mbostock/d3/wiki/Gallery)
* [Enter, Update, Exit 패턴](http://bost.ocks.org/mike/join/)
