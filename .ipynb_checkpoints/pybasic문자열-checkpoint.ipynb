###  python 문자열 다루는 함수 ( 추가)

- s.upper() → 문자열 s를 모두 대문자로 변경.
- s.lower() → 문자열 s를 모두 소문자로 변경.
- s.swapcase() → 문자열 s의 대문자는 소문자로, 소문자는 대문자로 각각 바꾸어 준다.
- s.count(x) → 문자열 s 중 문자 x와 일치하는 것의 개수를 반환(※ 이때 x는 문자 한 개 일수도, 문자열 일 수도 있다.)
- s.find(x) →  문자열 s 중 문자 x가 처음으로 나온 위치를 반환. 없으면 '-1'을 반환
- s.index(x) → 문자열 s 중 문자 x가 처음으로 나온 위치를 반환. 없으면 에러 발생.
- s.join(s) → Z라는 문자열의 각각의 요소 문자 사이에 문자열 s를 삽입
- s.strip() → 문자열 s의 양쪽 공백을 모두 제거
- s.lstrip() → 문자열 s의 왼쪽 공백을 모두 제거
- s.rstrip() → 문자열 s의 오른쪽 공백을 모두 제거
- s.replace(r, t) → 문자열 s의 r라는 문자열을 t이라는 문자열로 치환
- s.split(' ') → 문자열 s를 ' '(공백)을 구분자로 하여 나누어 준다. 반환된 결과는 리스트 이다. 

```python
#문자열 변수 선언
s="The Python Programming Language!!!"
```

```python
s.upper() # 모든 문자열을 대문자로 변경 
```

```python
s.lower() # 모든 문자열을 소문자로 변경
```

```python
s.swapcase()  # 대문자 -> 소문자로 소문자 -> 대문자로
```

```python
P=s.count('P')  #
p=s.count('p')
print('P: %d , p: %d' % (P, p))  # 문자열 카운터 가능 대소문자 구분
```

```python
s.find('P')  # 문자 'P' 가 처음 나온 위치값을 반환 
#문자열: "The Python Programming Language!!!"
#-------->0123456789012345678901234567890123-
```

```python
s.index('P')  # 문자 'P' 가 처음 나온위치값을 반환 
```

```python
#s.join('P')  #확인 필요 
s.strip()
s.lstrip()
s.rstrip()
```

```python
s.replace('P', 'p') ## 'P'을 'p'로 치환  
```

```python
l=s.split(' ')  # s라는 문자열을 공백으로 나눈다. 결과는 list로 반환
print(l)
type(l)
```
