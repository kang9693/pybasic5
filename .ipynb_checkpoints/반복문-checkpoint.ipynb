### while 문의 기본구조 

- 반복해서 문장을 수행해야 할 경우 while 문을 사용
- while문은 반복문이라고도 한다. 
- 조건을 만족할때 까지 반복 한다.
- While 문의 문장 구조 

```python
#while 조건문:
#    수행할 문장1
#    수행할 문장2
#   수행할 문장3
    
#while 문의 조건문이 참인 동안에 while 문 아래에 속한는 문장들이 반복해서 수행된다.
```

```python
# "열번 찍어 안 넘어가는 나무 없다.  " 를 구현 해보자.
```

```python
treeHit =0 
while treeHit < 10:
    treeHit=treeHit + 1
    print("나무를 %d번 찍어습니다. " % treeHit)
    if treeHit ==10:
        print("나무 넘어갑니다.!!!")
```

```python
# for을 이용하여 구현 해보자.!!!

treeHit =0 
#while treeHit < 10:
for treeHit in range(0,10):
    treeHit=treeHit + 1  
    print("나무를 %d번 찍어습니다. " % treeHit)
    if treeHit ==10:
        print("나무 넘어갑니다.!!!!!!!")
```
