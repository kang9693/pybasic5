```python
#### readline()함수 이용하기 
```

```python
#readline.py
f =open("test.txt",'r')
line = f.readline()
print(line)
```

```python
## 위처럼 한줄만 출력 
```

```python
f =open("test.txt",'r')
while True:
    line = f.readline()
    if not line: break        
    #print(line.split("\n"))
    print(line.strip('\n'))
    #print(line.split("\n"))
f.close()
```

```python
f =open("test.txt",'r')
while True:
    line = f.readline()
    if not line: break        
    print(line.split(" "))    
    
f.close()
```

```python
while 1:
    data=input()
    if not data: break
    print(data)
```

```python
#### readlines()함수를 이용하기
```

```python
f =open("test.txt",'r')
while True:
    line = f.readlines()   ## 결과를 리턴 하는 방법이 바뀐다.
    if not line: break        
    print(line)
f.close()
```

```python
### 파일에 새로운 내용 추가하기 
```

```python
## adddata.py
f =open("test.txt",'a')   ## 파일 읽기 옵션 a 
for i in range(11,20):  ## 11부터 19까지 i에 대입
    data = "%d번째 줄입니다.\n" % i 
    f.write(data)
f.close()
```

```python
f =open("test.txt",'r')
data = f.read()
print(data)
f.close()
```

```python
## adddata.py
f =open("test.txt",'a')   ## 파일 읽기 옵션 a 
for i in range(10,12):  ## 11부터 19까지 i에 대입
    data = "%d번째 줄입니다.\n" % i 
    f.write(data)
f.close()
```

```python
f =open("test.txt",'r')
data = f.read()
print(data)
f.close()  # 이런게 해도 끝에 추가 된다. 
```

```python
##with 문과 함께 사용하기
f =open("test.txt",'r')
data = f.read()
print(data)
f.close()  # 이런게 해도 끝에 추가 된다. 

```

```python
with open("foo.txt","w") as f:
    f.write("Lift is too short, you need python")
    
#with 문을 사용하면 블록을 벋어 나는 순간 자동으로 f.close() 되어 편리하다. 
```
