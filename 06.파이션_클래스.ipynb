### 05장 파이썬 날개달기

```python
## **kwargs 와같이 변수 사용법의 예

# **kwargs
# 파라미터 명을 같이 보낼 수 있다. kwargs는 딕셔너리 형태로 전달된다.
## **kwargs 사용법

def dic_param2(**kwargs):
    print ("kwargs:",kwargs)
    print ("key:",kwargs.keys())
    print ("value:",kwargs.values())
 
    for name, value in kwargs.items():
        print ("%s : %s" % (name, value))
 
dic_param2(first = 'a', second = 'b', third = 'c', fourth = 'd')
```

- **kwargs 사용법 (dic  )

```python
#두개를 같이 쓰는 경우는?
def args_param3(*args, **kwargs):
    print (args)
    print (kwargs)
```

```python
args_param3('a', 'b')
```

```python
args_param3(third = 'c', fourth = 'd')
```

```python
args_param3('a', 'b', third = 'c', fourth = 'd')
```

- 간단한 응용

```python
def args_param4(a, b, c):
    print (a, b, c)
```

```python
p = ['a', 'b', 'c']
args_param4(*p)
```

```python
p2 = {'c' : '1', 'a' : '2', 'b' : '3'}
args_param4(**p2)
```

- 클래스는 도대체 왜 필요한가?

```python
#가장 많이 사용하는 프로그래밍 언어 중 하나인 C 언어에는 클래스가 없다. 
#이 말은 굳이 클래스 없이도 프로그램을 충분히 만들 수 있다는 말과도 같다. 
#파이썬으로 잘 만들어진 프로그램들을 살펴보아도 클래스를 이용하지 않고 작성된 것들이 상당히 많다. 
#클래스는 지금까지 공부한 함수나 자료형처럼 프로그램 작성을 위해 꼭 필요한 요소는 아니다. 
#하지만 이러한 이유로 클래스에 접근하기 매우 까다로운 측면이 있다. 왜 이걸 굳이 클래스로 해야하지? 
#라는 의문이 항상 뒤따르기 때문이다.
#하지만 프로그램 작성시 클래스를 적재적소에 이용하면 프로그래머가 얻을 수 있는 이익은 상당하다.
#예제를 통해 한번 생각해 보자.
#여러분 모두 계산기를 사용해 봤을 것이다. 계산기에 3이라는 숫자를 입력하고 + 기호를 입력한 후 4를 입력하면 결과값으로 7을 보여준다.
#다시 한 번 + 기호를 입력한 후 3을 입력하면 기존 결과값 7에 3을 더해 10을 보여준다.
#즉, 계산기는 이전에 계산된 결과값을 항상 메모리 어딘가에 저장하고 있어야 한다.
```

```python
result = 0

def adder(num):
    global result  #global 변수란. 
    result += num
    return result

print(adder(3))
print(adder(4))
```

```python
result1 = 0
result2 = 0

def adder1(num):
    global result1
    result1 += num
    return result1

def adder2(num):
    global result2
    result2 += num
    return result2

print(adder1(3))
print(adder1(4))
print(adder2(3))
print(adder2(7))
```

```python
class Calculator:
    def __init__(self):
        self.result = 0

    def adder(self, num):
        self.result += num
        return self.result

cal1 = Calculator()
cal2 = Calculator()

print(cal1.adder(3))
print(cal1.adder(4))
print(cal2.adder(3))
print(cal2.adder(7))
```
