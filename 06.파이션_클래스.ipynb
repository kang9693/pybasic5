### 05장 파이썬 날개달기

```python
## **kwargs 와같이 변수 사용법의 예

# **kwargs
# 파라미터 명을 같이 보낼 수 있다. kwargs는 딕셔너리 형태로 전달된다.
## **kwargs 사용법

def dic_param2(**kwargs):
    print ("kwargs:",kwargs)
    print ("key:",kwargs.keys())
    print ("value:",kwargs.values())
 
    for name, value in kwargs.items():
        print ("%s : %s" % (name, value))
 
dic_param2(first = 'a', second = 'b', third = 'c', fourth = 'd')
```

- **kwargs 사용법 (dic  )

```python
#두개를 같이 쓰는 경우는?
def args_param3(*args, **kwargs):
    print (args)
    print (kwargs)
```

```python
args_param3('a', 'b')
```

```python
args_param3(third = 'c', fourth = 'd')
```

```python
args_param3('a', 'b', third = 'c', fourth = 'd')
```

- 간단한 응용

```python
def args_param4(a, b, c):
    print (a, b, c)
```

```python
p = ['a', 'b', 'c']
args_param4(*p)
```

```python
p2 = {'c' : '1', 'a' : '2', 'b' : '3'}
args_param4(**p2)
```
